```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Define the range for X1 and X2
X1 = np.linspace(-3, 3, 400)
X2 = np.linspace(-1, 5, 400)
X1, X2 = np.meshgrid(X1, X2)

# Define the curve based on the equation
curve = (1 + X1)**2 + (2 - X2)**2 - 4

# Plot for question (a)
sns.heatmap(curve > 0, cbar=False, cmap='coolwarm', ax=axes[0])
axes[0].set_title('Decision Boundary: (1 + X1)^2 + (2 - X2)^2 > 4')
axes[0].set_xlabel('X1')
axes[0].set_ylabel('X2')

# Plot for question (b)
sns.heatmap(curve > 0, cbar=False, cmap='coolwarm', ax=axes[1])
sns.heatmap(curve <= 0, cbar=False, cmap='coolwarm', ax=axes[1], alpha=0.4, annot=True, fmt='d', annot_kws={"size": 12})
axes[1].set_title('Blue and Red Regions')
axes[1].set_xlabel('X1')
axes[1].set_ylabel('X2')

# Set titles for the entire figure
# fig.suptitle('Non-Linear Decision Boundary and Region Classification', fontsize=16)

# # Show the plots
# plt.tight_layout()
# plt.show()
```
